```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img_dark = np.array(Image.open("./hdr_images/night_dark.jpg"))
img_light = np.array(Image.open("./hdr_images/night_light.jpg"))

l, c, p = img_dark.shape
```

```python
img_hdr = np.zeros(shape=img_dark.shape, dtype=np.float64)
for i in range(l):
    for j in range(c):
        r = (img_dark[i, j, 0]*0.5 + 0.5*img_light[i, j, 0])/255
        g = (img_dark[i, j, 1]*0.5 + 0.5*img_light[i, j, 1])/255
        b = (img_dark[i, j, 2]*0.5 + 0.5*img_light[i, j, 2])/255
        
        img_hdr[i, j, 0] = r
        img_hdr[i, j, 1] = g
        img_hdr[i, j, 2] = b
```

```python
plt.figure(figsize=(25, 25))
plt.subplot(3, 2, 1)
plt.imshow(img_dark)
plt.subplot(3, 2, 2)
plt.imshow(img_light)
plt.subplot(3, 2, 3)
plt.imshow(img_hdr)


plt.imshow(img_hdr, cmap='gray')
# plt.savefig("night_hdr_comparation.png")
```
