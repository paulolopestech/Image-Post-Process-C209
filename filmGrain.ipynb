```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

image = np.array(Image.open('./grain_images/beach.jpg'))[:, :, :3]
noise = np.array(Image.open('noise.jpeg'))[:, :, :3]

(lb, cb, pb) = image.shape

grain = np.zeros((lb, cb, pb), dtype=np.uint8)
```

```python
gw = 1
for i in range(lb):
    for j in range(cb):
        if(image[i, j, 0] + gw*noise[i, j, 0] > 255):
            grain[i, j, 0] = 255
        else:
            grain[i, j, 0] = image[i, j, 0] + gw*noise[i, j, 0]
        
        if(image[i, j, 1] + gw*noise[i, j, 1] > 255):
            grain[i, j, 1] = 255
        else:
            grain[i, j, 1] = image[i, j, 1] + gw*noise[i, j, 1]
        
        if(image[i, j, 2] + gw*noise[i, j, 2] > 255):
            grain[i, j, 2] = 255
        else:
            grain[i, j, 2] = image[i, j, 2] + gw*noise[i, j, 2]
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(3, 2, 1)
plt.imshow(image)
plt.subplot(3, 2, 2)
plt.imshow(grain)
# plt.savefig("beach_grain_comparation.png")
```
